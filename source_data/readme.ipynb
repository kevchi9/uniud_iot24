```python
import pickle
import json
import math
```

```python
with open('batch0.pickle', 'rb') as logfile:
    data = pickle.load(logfile)

times = []

#La chiave più esterna è un tempo assoluto (tempo calcolato da python, non dal simulatore)
for time in data:
    times.append(time)
print(times)
print(len(times))

# conta il numero di occorrenze in un secondo
def count_occurrences_in_second(data, times):
    occurrences = 0
    for key in data:
        key = math.trunc(key)
        if key == times:
            occurrences += 1
    return occurrences

for i in range(1715776778, 1715776820):  # range in secondi
    print(count_occurrences_in_second(data, i))
```
